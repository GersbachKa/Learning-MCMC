```python
import os
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
print("Gaussian one parameter:")
with open("gaussianOneParam_0.txt",'r') as f:
    read = csv.reader(f);
    data = np.array(list(read))[:,0].astype(np.float)
plt.plot(data)
plt.show()
plt.hist(data,100,density=True)
plt.show()
```

```python
print("Gaussian two parameter:")
with open("gaussianTwoParam_0.txt",'r') as f:
    read = csv.reader(f);
    data0 = np.array(list(read))[:,0:2].astype(np.float)
    
plt.plot(data0[:,0])
plt.plot(data0[:,1],'r')
plt.show()
burnedData = data0[int(len(data0)*0.25):] 
plt.hist(burnedData[:,0],100,density=True)#mu
plt.show()
plt.hist(burnedData[:,1],100,density=True)#Sigma
plt.show()

plt.hist2d(burnedData[:,0],burnedData[:,1],bins=[100,100],density=True)
plt.show()
```
